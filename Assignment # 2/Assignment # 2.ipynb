```python
import pandas as pd

df = pd.read_csv('pokemon.csv')
df.head()
```

```python
# Means Of Hp and Speed Columns
hp_mean = df['HP'].mean()
speed_mean = df['Speed'].mean()

# Defining Methods for Hp_high_low and Speed_high_low
def hp_high_low(hp):
    if hp < hp_mean:
        return 'Low HP'
    else:
        return 'High HP'
    
def speed_high_low(speed):
    if speed < speed_mean:
        return 'Low Speed'
    else:
        return 'High Speed'

# Creating HP_High_low and Speed_High_Low columns
df['HP_High_Low'] = df['HP'].apply(hp_high_low)
df['Speed_High_Low'] = df['Speed'].apply(speed_high_low)
df.head(10)
```
